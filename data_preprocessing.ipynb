## Clubs Preprocessing

```python
# Extract Year from date
df_player_valuations['date'] = pd.to_datetime(df_player_valuations['date'])
df_player_valuations['year'] = df_player_valuations['date'].dt.year

# Get the static club info
static_club_info = df_clubs[['club_id', 'name', 'domestic_competition_id', 'squad_size', 
                             'average_age', 'foreigners_number', 'foreigners_percentage', 
                             'national_team_players', 'stadium_name', 'stadium_seats', 
                             'net_transfer_record', 'coach_name', 'filename', 'url']]

# Get the highest market value of each player per year
max_player_valuations = df_player_valuations.groupby(['player_id', 'year'])['market_value_in_eur'].max().reset_index()

# Create a complete dataframe with all player-year combinations
all_years = pd.DataFrame({'year': range(df_player_valuations['year'].min(), df_player_valuations['year'].max() + 1)})
all_players = df_player_valuations[['player_id']].drop_duplicates()
player_years = all_players.merge(all_years, how='cross')  # Cartesian product to get all combinations

# Merge with max valuations and forward-fill missing values
player_valuations_filled = player_years.merge(max_player_valuations, on=['player_id', 'year'], how='left')
player_valuations_filled['market_value_in_eur'] = player_valuations_filled.groupby('player_id')['market_value_in_eur'].ffill()

# Merge back to get club information
df_max_valuations = df_player_valuations[['player_id', 'year', 'current_club_id']].drop_duplicates()
df_max_valuations = player_valuations_filled.merge(df_max_valuations, on=['player_id', 'year'], how='left')

# Sum player values per club per year
club_market_values = df_max_valuations.groupby(['current_club_id', 'year'])['market_value_in_eur'].sum().reset_index()

# Rename column
club_market_values.rename(columns={'market_value_in_eur': 'total_market_value'}, inplace=True)

# Merge with club names
club_market_values = club_market_values.merge(df_clubs[['club_id', 'name']], left_on='current_club_id', right_on='club_id', how='left')

# Drop redundant columns
club_market_values.drop(columns=['current_club_id', 'club_id'], inplace=True)

# Sort values
club_market_values.sort_values(by=['name', 'year'], inplace=True)
```

```python
# Save the processed data to CSV files
static_club_info.to_csv("preprocessed_data/static_club_info.csv", index=False)
club_market_values.to_csv("preprocessed_data/club_market_value_over_time.csv", index=False)
```
